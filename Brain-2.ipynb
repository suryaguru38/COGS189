# Load Data

 ### First Sheet "Consolidated_ERS_ERD_demographi"

```python
excel_file = 'BrainE_MLmodel_specs_consolidated_v3.xlsx'
all_features_in_first_sheet = pd.read_excel(excel_file, sheet_name="Consolidated_ERS_ERD_demographi")
```

### Select those features
"ERD1", "ERD2", "ERD3", "ERD5", "ERS1", "age", "gender", "race", "ses", "childhoodevents", "sleep", "mental_health"

```python
selected_in_features_First_sheet = all_features_in_first_sheet.loc[:, ["ERD1", "ERD2", "ERD3", "ERD5", "ERS1", "age", "gender", "race", "ses", "childhoodevents", "sleep", "mental_health"]]
```

### Select First 84 row 
Cause from row number 85 to 98 don't have values for mental health and i will effect the results

```python
selected_in_features_First_sheet = selected_in_features_First_sheet[:84]
```

# Draw Correlation Matrix

```python
selected_in_features_First_sheet.corr().style.background_gradient(cmap='coolwarm')
```
